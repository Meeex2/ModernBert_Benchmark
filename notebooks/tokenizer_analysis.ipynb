```python
from transformers import AutoTokenizer, ModernBertModel
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("answerdotai/ModernBERT-base")
model = ModernBertModel.from_pretrained("answerdotai/ModernBERT-base")

# Define the pooling function
def pool_encoder_output(last_hidden_states, pooling_type='average'):
    if pooling_type == 'average':
        pooled_output = torch.mean(last_hidden_states, dim=1)
    elif pooling_type == 'max':
        pooled_output, _ = torch.max(last_hidden_states, dim=1)
    elif pooling_type == 'min':
        pooled_output, _ = torch.min(last_hidden_states, dim=1)
    else:
        raise ValueError("Invalid pooling type. Choose from 'average', 'max', or 'min'.")
    return pooled_output

# Tokenize the input and get the model output
inputs = tokenizer("Hello, my dog is cute", return_tensors="pt")
outputs = model(**inputs)

# Get the last hidden states from the model output
last_hidden_states = outputs.last_hidden_state

# Perform pooling on the last hidden states
pooled_output = pool_encoder_output(last_hidden_states, pooling_type='average')

# Print the pooled output
print(pooled_output)
```

```python
from transformers import AutoTokenizer

# List of colors in RGB format
colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    # Load the tokenizer
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

    # Tokenize the sentence
    token_ids = tokenizer(sentence).input_ids

    # Print each token with a color from the colors_list
    for idx, t in enumerate(token_ids):
        # Get the color for the current token
        color = colors_list[idx % len(colors_list)]

        # Decode the token and print it with the corresponding color
        print(
            f'\x1b[0;30;48;2;{color}m' +
            tokenizer.decode([t]) +
            '\x1b[0m',
            end=' '
        )
    print()  # Print a newline at the end

# Example usage with a code snippet
code_snippet = """
def hello_world():
    print("Hello, world!")
"""
tokenizer_name = "answerdotai/ModernBERT-base"
show_tokens(code_snippet, tokenizer_name)
```
