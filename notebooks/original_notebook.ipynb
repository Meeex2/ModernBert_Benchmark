```python
from transformers import AutoTokenizer

# List of colors in RGB format
colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    # Load the tokenizer
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

    # Tokenize the sentence
    token_ids = tokenizer(sentence).input_ids

    # Print each token with a color from the colors_list
    for idx, t in enumerate(token_ids):
        # Get the color for the current token
        color = colors_list[idx % len(colors_list)]

        # Decode the token and print it with the corresponding color
        print(
            f'\x1b[0;30;48;2;{color}m' +
            tokenizer.decode([t]) +
            '\x1b[0m',
            end=' '
        )
    print()  # Print a newline at the end

# Example usage with a code snippet
code_snippet = """
def hello_world():
    print("Hello, world!")
"""
tokenizer_name = "answerdotai/ModernBERT-base"
show_tokens(code_snippet, tokenizer_name)
```
